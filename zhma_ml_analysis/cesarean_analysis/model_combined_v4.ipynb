```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from utils import *
```

```python
file_names = ['static',
              'first_trimster',
              'second_trimster',
              'one_month_before_delivery',
              '72h_before_delivery',
              '24h_before_delivery',
              '12h_before_delivery',
              '3h_before_delivery',
              '1h_before_delivery',
              'at_delivery']
```

```python
auc_test_scores = []
f1_test_scores = []

for file_name in file_names:

    print(f'doing {file_name}')
    
    if file_name == 'static':
        dat = pd.read_parquet('dat_static.parquet')
    else:
        dat = pd.read_parquet(f'dat_track_{file_name}.parquet')

    fill_nans(dat)
    tmp = dat.drop(columns=['BIRTHID', 'preg_start_date']).dropna()
    tmp = tmp[tmp.label>=0].copy()
    tmp.shape
    
    clf, X_test, y_test, auc_test_score, f1_test_score = get_fit(tmp)
    auc_test_scores.append(auc_test_score)
    f1_test_scores.append(f1_test_score)

    fig, fig2 = get_shap(clf.best_estimator_, X_test, y_test);
    fig.savefig(f"{file_name}_shap.pdf");
    fig2.savefig(f"{file_name}_importance.pdf", bbox_inches='tight');
```

```python
plt.plot(f1_test_scores, 'o-');
plt.ylabel('f1 score')
plt.xticks(ticks=range(len(f1_test_scores)),labels=file_names, rotation=70);
```
